```python
import re
```

```python
with open('./sticker.csv', 'r') as f:
    data = f.read()

data.split(',')
```

```python
regex = re.compile('^(\d+\.\d+)([A-Za-z_]+)$')
regex2 = re.compile('^(\d+\.\d+)([A-Z\-]+\d+)$')

result = []

for string in data.split(','):
    match = regex.match(string)
    match2 = regex2.match(string)
    
    if string != 'nan' and 'nan' in string:
        temp = string.split('nan')
        for x in temp:
            if x == '':
                result.append('nan')
            else:
                result.append(x)
    elif match:
        result.append(match.group(1))
        result.append(match.group(2))
    elif match2:
        result.append(match2.group(1))
        result.append(match2.group(2))
    elif string == 'currentWTW':
        result.append('current')
        result.append('WTW')
    else:
        result.append(string)

result
```

```python
with open('./sticker2.csv', 'a') as f:
    for i in range(0, len(result), 3):
        f.write(','.join(result[i:i+3]) + '\n')
```
